```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = .5
v = .3
l = 100
s = 0.7
x = np.linspace(0, 1, 100)
y = (s / (1 + np.exp(-l*(x-h)))) + v
```

```python
plt.plot(x, y)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title("Sigmoid for Dublin with Shift")
plt.xlabel("Percent of dublins x-axis")
plt.ylabel("Intensity Modulation")
```
